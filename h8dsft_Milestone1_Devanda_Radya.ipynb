#### Analisis dan perhitungan

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway
```

##### Statistik Deskriptif (Harga Produk)

```python
# Hitung statistik deskriptif
mean_price = df['Purchase Amount (USD)'].mean() # Menghitung AVG (mean) pada kolom 'Purchase Amount (USD)'.
median_price = df['Purchase Amount (USD)'].median() # Menghitung nilai tengah (median) pada kolom 'Purchase Amount (USD)'.
mode_price = df['Purchase Amount (USD)'].mode()[0] # Mengambil nilai modus (nilai yang paling sering muncul) pada kolom 'Purchase Amount (USD)'. Indeks [0] digunakan karena mode() mengembalikan Series.
std_dev = df['Purchase Amount (USD)'].std() # Menghitung standar deviasi (ukuran sebaran data) pada kolom 'Purchase Amount (USD)'.

print(f"Rata-rata Harga: {mean_price}")
print(f"Median Harga: {median_price}")
print(f"Modus Harga: {mode_price}")
print(f"Standard Deviasi: {std_dev}")

# Visualisasi distribusi harga dengan garis referensi
plt.figure(figsize=(8,6)) # Mengatur ukuran figure untuk visualisasi (8 lebar, 6 tinggi).
# Membuat histogram distribusi harga dengan 20 bin dan garis KDE (Kernel Density Estimation) untuk memperlihatkan bentuk distribusi.
sns.histplot(df['Purchase Amount (USD)'], bins=20, kde=True, color='purple')

# Menambahkan garis vertikal merah putus-putus pada posisi avg.
plt.axvline(mean_price, color='red', linestyle='--', label='Mean')
# Menambahkan garis vertikal hijau pada posisi median.
plt.axvline(median_price, color='green', linestyle='-', label='Median')
# Menambahkan garis vertikal biru titik-putus pada posisi modus.
plt.axvline(mode_price, color='blue', linestyle='-.', label='Mode')
plt.title('Distribusi Harga Produk')
plt.xlabel('Harga (USD)') # Sumbu x
plt.ylabel('Frekuensi') # Sumbu Y
plt.legend()
plt.show()
```

1. Statistik Deskriptif (Central Tendency dan Measure of Variance):
    - Mean: rata-rata pengeluaran pembeli
    - Median: mengidentifikasi nilai tengah
    - Mode: menunjukkan nilai yang paling sering terjadi
    - std: mengukur seberapa jauh nilai dari rata-rata

2. Visualisasi histogram dan garis KDE (Kernel Density Estimation):
    - Histogram untuk memperlihatkan frekuensi pengeluaran pada rentang tertentu
    - Garis (Mean, Median, Mode) membantu membandingkan hubungan antara tiga metrik utama
    - Garis KDE memberikan gambaran bentuk distribusi data (normal, skewed, atau lainnya)

Alasan Menggunakan Statistik Ini:
- Central Tendency (Mean, Median, Mode):
    - Dapat mengetahui tren pengeluaran pelanggan
    - Dapat membantu menentukan stategi target promosi

- Measure of Variance (Standard Deviation):
    - Banyaknya variasi mungkin menunjukkan segmen pelanggan yang sangat berbeda, yang memerlukan pendekatan berbeda

- Distribution:
    - Distribusi memberikan konteks lebih luas. misal, apakah pengeluaran pelanggan tersebar secara merata, dominan di kelompok tertentu.

##### Statistik Inferensial (Uji ANOVA)

```python
# Kelompok usia (rentang usia: <25, 25-40, >40)
# Membuat kolom baru 'Age Group' dengan mengelompokkan usia (kolom 'Age') ke dalam 3 kategori:
df['Age Group'] = pd.cut(df['Age'], bins=[0, 25, 40, 100], labels=['<25', '25-40', '>40'])
# - "<25" untuk usia 0-25
# - "25-40" untuk usia 26-40
# - ">40" untuk usia di atas 40
# `bins` menentukan rentang usia, sedangkan `labels` memberikan label kategori.

# Group berdasarkan kolompok usia
# Mengambil data kolom 'Purchase Amount (USD)' untuk kelompok usia "<25".
group1 = df[df['Age Group'] == '<25']['Purchase Amount (USD)']
# Mengambil data kolom 'Purchase Amount (USD)' untuk kelompok usia "25-40".
group2 = df[df['Age Group'] == '25-40']['Purchase Amount (USD)']
# Mengambil data kolom 'Purchase Amount (USD)' untuk kelompok usia ">40".
group3 = df[df['Age Group'] == '>40']['Purchase Amount (USD)']

# Uji ANOVA
anova_result = f_oneway(group1, group2, group3)
# Melakukan uji ANOVA satu arah untuk membandingkan avg pengeluaran ('Purchase Amount (USD)')
# antar tiga kelompok usia (group1, group2, dan group3).
# Fungsi dari `f_oneway()` adalah mengembalikan F-statistic dan p-value.

print("Hasil Uji ANOVA:")
# Menampilkan hasil uji ANOVA, yaitu nilai F-statistic dan p-value dengan 2 dan 4 angka dibelakang koma.
print(f"F-statistic: {anova_result.statistic:.2f}, p-value: {anova_result.pvalue:.4f}")

# Interpretasi
# Jika p-value < 0.05, maka terdapat perbedaan signifikan dalam pengeluaran antar kelompok usia.
if anova_result.pvalue < 0.05:
    print("Terdapat perbedaan signifikan dalam pengeluaran berdasarkan kelompok usia.")
else:
    print("Tidak terdapat perbedaan signifikan dalam pengeluaran berdasarkan kelompok usia.")

# Hitung rata-rata pengeluaran untuk masing-masing kelompok usia
avg_usia = df.groupby('Age Group')['Purchase Amount (USD)'].mean()

# Buat bar plot untuk rata-rata pengeluaran per kelompok usia
plt.figure(figsize=(8, 6))  # Mengatur ukuran figure
avg_usia.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'], edgecolor='black')

# Menambahkan detail pada grafik
plt.title('Rata-rata Pengeluaran Berdasarkan Kelompok Usia', fontsize=14)
plt.xlabel('Kelompok Usia', fontsize=12)
plt.ylabel('Rata-rata Pengeluaran (USD)', fontsize=12)
plt.xticks(rotation=0)  # Menampilkan label x secara horizontal
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Menambahkan grid horizontal
plt.show()
```

ANOVA (Analysis of Variance) adalah metode statistik untuk membandingkan rata-rata antar tiga atau lebih kelompok. Pada kasus ini, ANOVA digunakan untuk menganalisis perbedaan rata-rata pengeluaran (Purchase Amount (USD)) berdasarkan kelompok usia:
- (<25 tahun)
- (25â€“40 tahun)
- (>40 tahun)

Alasan Menggunakan ANOVA

1. Efisiensi Analisis: Membandingkan tiga kelompok sekaligus tanpa perlu melakukan banyak uji t-test.
2. Mengidentifikasi Pola: ANOVA membantu melihat apakah faktor usia memengaruhi pengeluaran pelanggan.
3. Mendukung Strategi Bisnis: Hasilnya digunakan untuk menyesuaikan strategi pemasaran atau promosi sesuai dengan kelompok usia.

##### TOP 10 Produk Terlaris

```python
# Hitung produk terlaris
# Menghitung jumlah kemunculan pada setiap produk pada kolom 'Item Purchased' menggunakan value_counts().
# Mengambil 10 produk dengan frekuensi tertinggi menggunakan head(10).
# Hasil berupa Series dengan produk sebagai indeks dan jumlah pembelian sebagai nilai.
top10_produk = df['Item Purchased'].value_counts().head(10)

# Visualisasi produk terlaris
# Mengatur ukuran figure (lebar=11, tinggi=6) agar visualisasi lebih jelas.
plt.figure(figsize=(11,6))

# Membuat barplot menggunakan Seaborn.
# - sumbu X: Nilai dari top10_produk (jumlah pembelian).
# - sumbu Y: Indeks dari top10_produk (nama produk).
# - palette='Blues_d' adalah gradasi warna biru.
sns.barplot(x=top10_produk.values, y=top10_produk.index, palette='Blues_d')
plt.title('Top 10 Produk Terlaris')
plt.xlabel('Jumlah Pembelian')
plt.ylabel('Produk')
plt.show()
```

##### Rata-rata Pengeluaran Berdasarkan Lokasi

```python
# Hitung avg pengeluaran berdasarkan lokasi
# `df.groupby('Location')`: pengelompokan data pada kolom 'Location'.
# `['Purchase Amount (USD)'].mean()`: Menghitung avg pada setiap lokasi.
# `.sort_values(ascending=False)`: Mengurutkan hasil avg dari yang terbesar ke terkecil.
# Tambahan opsional: Jika ingin TOP 10 lokasi Tertinggi, tambahkan `.head(10)`.
pengeluaran_lokasi = df.groupby('Location')['Purchase Amount (USD)'].mean().sort_values(ascending=False).head(10)

# Visualisasi
plt.figure(figsize=(11,6))
# Membuat barplot:
# - `x=pengeluaran_lokasi.values`: Nilai avg sebagai sumbu X.
# - `y=pengeluaran_lokasi.index`: Nama lokasi sebagai sumbu Y.
# - `palette='Greens_d'`: Memberikan warna hijau.
sns.barplot(x=pengeluaran_lokasi.values, y=pengeluaran_lokasi.index, palette='Greens_d')
plt.title('Rata-rata Pengeluaran Berdasarkan Lokasi')
plt.xlabel('Rata-rata Pengeluaran (USD)')
plt.ylabel('Lokasi')
plt.show()
```

##### Pembelian Berdasarkan Waktu (Season)

```python
# Hitung jumlah pembelian per musim
# `df['Season']`: Mengambil kolom 'Season' dari DataFrame.
# `value_counts()`: Menghitung jumlah kemunculan (frekuensi) setiap kategori musim.
purchase_season = df['Season'].value_counts()

# Urutkan berdasarkan musim (jika musim adalah kategori tertentu)
# Pastikan musim diurutkan secara logis, misalnya: ['Winter', 'Spring', 'Summer', 'Fall']
ordered_seasons = ['Winter', 'Spring', 'Summer', 'Fall']
purchase_season = purchase_season.reindex(ordered_seasons)

# Visualisasi
# Mengatur ukuran plot dengan lebar 8 dan tinggi 6 agar visualisasi lebih jelas.
plt.figure(figsize=(8,6))
# Membuat line plot:
# - `x=purchase_season.index`: Sumbu X menampilkan musim.
# - `y=purchase_season.values`: Sumbu Y menampilkan jumlah pembelian per musim.
# - `marker='o'`: Menambahkan penanda lingkaran pada setiap titik data untuk memperjelas nilai.
sns.barplot(x=purchase_season.index, y=purchase_season.values, palette='viridis')
plt.title('Jumlah Pembelian Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Jumlah Pembelian')
plt.show()
```

##### Pembelian Berdasarkan Usia dan Gender:

```python
# Membuat kolom 'Age Group' untuk membagi usia dalam selisih 2 tahun
# `pd.cut()`: Membagi nilai kolom 'Age' ke dalam kelompok selisih yang telah ditentukan.
# `bins=range(0, 101, 2)`: Membuat rentang usia dari 0 hingga 100 dengan selisih 2 tahun.
# `right=False`: Menentukan bahwa batas kanan tidak termasuk kelompok.
# `labels=[f'{i}-{i+1}' for i in range(0, 100, 2)]`: Membuat label untuk setiap kelompok usia.
df['Age Group'] = pd.cut(df['Age'], bins=range(0, 101, 2), right=False, labels=[f'{i}-{i+1}' for i in range(0, 100, 2)])

# Visualisasi jumlah pembelian berdasarkan usia dan gender
# Mengatur ukuran plot dengan lebar 25 dan tinggi 10 untuk memastikan semua kelompok usia terlihat dengan jelas.
plt.figure(figsize=(25, 10))
# Membuat histogram:
# - `x='Age Group'`: Sumbu X menampilkan usia.
# - `hue='Gender'`: Membagi histogram berdasarkan ('Gender'), dengan warna berbeda.
# - `multiple='stack'`: Menumpuk histogram untuk membedakan jumlah gender pada kelompok usia.
# - `bins=len(df['Age Group'].unique())`: Jumlah histogram disesuaikan dengan jumlah kelompok usia unik.
# - `palette='coolwarm'`: Membedakan warna setiap jenis kelamin.
sns.histplot(data=df, x='Age Group', hue='Gender', multiple='stack', bins=len(df['Age Group'].unique()), palette='coolwarm')
plt.title('Distribusi Pembelian Berdasarkan Usia dan Jenis Kelamin')
plt.xlabel('Kelompok Usia')
plt.ylabel('Jumlah Pembelian')
plt.show()
```
